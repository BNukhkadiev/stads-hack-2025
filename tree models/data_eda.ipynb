### encode label as 1 and 0, 1 for anomal

```python
# encode label column to 1 for irregular an 0 for regular
df['label'] = df['label'].map({'regular': 0, 'anomal': 1})
```

```python
df.head(20)
```

```python
# Identify categorical columns
categorical_cols = ["WAERS", "BUKRS", "KTOSL", "PRCTR", "BSCHL", "HKONT"]
```

```python
# encode categorical variables
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store encoders for inverse transformation
```

```python
# frequency encode categorical variables
frequency_map = {}
for col in categorical_cols:
    frequency_map[col] = df[col].value_counts(normalize=False).to_dict()

for col in categorical_cols:
    df[col] = df[col].map(frequency_map[col])

print(df.shape)
df.head(20)
```

## 30 irregular anomalies that are not just unique categories

```python
# Loop over categorical columns and drop categories that appear only once
for col in categorical_cols:
    # Get the value counts of each category in the column
    value_counts = df[col].value_counts()
    
    # Find categories that appear only once
    categories_to_drop = value_counts[value_counts == 1].index
    
    # Drop these categories from the column
    df_hard_irregular = df[~df[col].isin(categories_to_drop)]

print(df_hard_irregular.shape)
df_hard_irregular.head(20)
```

```python
# frequency encode categorical variables
frequency_map = {}
for col in categorical_cols:
    frequency_map[col] = df_hard_irregular[col].value_counts(normalize=False).to_dict()

for col in categorical_cols:
    df_hard_irregular[col] = df_hard_irregular[col].map(frequency_map[col])

print(df_hard_irregular.shape)
df_hard_irregular.head(20)
```

```python
df_hard_irregular.to_csv('data/datathon_data_without_unique.csv', index=False)
```
